6. Visualizing the Distribution of Passengers by Embarked Port
Explanation: In this section, we analyze and visualize the distribution of passengers who embarked at each of the three ports: 'S', 'C', and 'Q'. We count the number of passengers at each port, calculate the percentages, and create a pie chart to represent this distribution visually.

```python
# Calculate the number of passengers for each Embarked port ('S', 'C', 'Q')
sclass = sum(training_data["Embarked"] == "S")
cclass = sum(training_data["Embarked"] == "C")
qclass = sum(training_data["Embarked"] == "Q")

# Create a list to store the passenger counts for each port
percentage = [sclass, cclass, qclass]

# Function to format the pie chart labels with both the count and percentage
def func(pct, allvalues):
    absolute = int(pct / 100. * sum(allvalues))
    return f"{absolute} ({pct:.1f}%)"

# Create the pie chart to visualize the distribution of passengers by Embarked port
plt.pie(percentage, labels=["S", "C", "Q"], 
        colors=["#05f5f5", "#e105f5", "#ff7f50"], shadow=True, 
        autopct=lambda pct: func(pct, percentage), startangle=180, explode=(0.1, 0.1, 0))

# Set the title of the pie chart
plt.title("Passenger Embarked Distribution")

# Display the pie chart
plt.show()


# Explanation of the Code:
#   Counting Passengers at Each Port:
#   sclass = sum(training_data["Embarked"] == "S"): This counts the number of passengers who boarded at port 'S' (Southampton).
#   cclass = sum(training_data["Embarked"] == "C"): This counts the number of passengers who boarded at port 'C' (Cherbourg).
#   qclass = sum(training_data["Embarked"] == "Q"): This counts the number of passengers who boarded at port 'Q' (Queenstown).
#   Percentage Calculation:
#   percentage = [sclass, cclass, qclass]: This list stores the counts for each port, which will later be used to calculate the percentages.
#   Function to Format Pie Chart Labels:
#   func(pct, allvalues): This function formats the pie chart labels to display both the absolute count and percentage for each port.
#   Plotting the Pie Chart:
#   plt.pie(): This function creates the pie chart with the specified labels, colors, and formatting. The explode argument separates the slices slightly for better visibility.
#   Displaying the Chart:
#   plt.title(): Sets the title of the chart to "Passenger Embarked Distribution".
#   plt.show(): Displays the pie chart.
```

18. Handling Missing Values in Numerical and Categorical Columns

```python
# Select columns with numerical data types (float64, int64)
numerical_cols = training_data.select_dtypes(include=['float64', 'int64']).columns

# Fill missing values in numerical columns with the mean of each column
training_data[numerical_cols] = training_data[numerical_cols].fillna(training_data[numerical_cols].mean())

# Select columns with categorical data types (object)
categorical_cols = training_data.select_dtypes(include=['object']).columns

# Fill missing values in categorical columns with the mode (most frequent value) of each column
for col in categorical_cols:
    training_data[col] = training_data[col].fillna(training_data[col].mode()[0])

# Print the number of missing values after preprocessing
print(f"Missing values after preprocessing : {training_data.isnull().sum()}")

# Explanation:
# - numerical_cols: This line selects all the columns with numerical data types (either int64 or float64).
# - The next line fills missing values in numerical columns with the mean of each respective column.
# - categorical_cols: This line selects all the columns with categorical data types (object).
# - The for loop fills missing values in each categorical column with the mode (most frequent value) of that column.
# - Finally, we print the number of missing values after preprocessing to verify that the missing values have been handled.
```

```python
# Select columns with numerical data types (float64, int64)
numerical_cols = testing_data.select_dtypes(include=['float64', 'int64']).columns

# Fill missing values in numerical columns with the mean of each column
testing_data[numerical_cols] = testing_data[numerical_cols].fillna(testing_data[numerical_cols].mean())

# Select columns with categorical data types (object)
categorical_cols = testing_data.select_dtypes(include=['object']).columns

# Fill missing values in categorical columns with the mode (most frequent value) of each column
for col in categorical_cols:
    testing_data[col] = testing_data[col].fillna(testing_data[col].mode()[0])

# Print the number of missing values after preprocessing
print(f"Missing values after preprocessing : {testing_data.isnull().sum()}")

# Explanation:
# - numerical_cols: This line selects all the columns with numerical data types (either int64 or float64).
# - The next line fills missing values in numerical columns with the mean of each respective column.
# - categorical_cols: This line selects all the columns with categorical data types (object).
# - The for loop fills missing values in each categorical column with the mode (most frequent value) of that column.
# - Finally, we print the number of missing values after preprocessing to verify that the missing values have been handled.
```

23. Defining Features and Target for the Model and Handling Categorical Variables

```python
# Define the features (X) and target (y) for the model
X = training_data[['Pclass', 'Sex', 'Age', 'SibSp', 'Parch', 'Fare']]
y = training_data['Survived']

# Explanation:
# X contains the independent variables (features) that will be used to predict the target variable.
# These features include 'Pclass', 'Sex', 'Age', 'SibSp', 'Parch', and 'Fare'.
# y contains the target variable, which is the 'Survived' column, representing the survival status (0 = No, 1 = Yes).

# Convert categorical variable 'Sex' to numerical values using one-hot encoding
X = pd.get_dummies(X, drop_first=True)

# Explanation:
# pd.get_dummies() is used to convert categorical variables into numerical values using one-hot encoding.
# 'Sex' is a categorical column with two possible values ('male' and 'female').
# One-hot encoding creates two new columns: 'Sex_male' and 'Sex_female', and it drops one column (using drop_first=True) to avoid multicollinearity.
# This ensures that the model can work with numerical data while preserving the categorical information.
```

24. Initializing and Training the Decision Tree Model

```python
# Initialize the Decision Tree model
model = DecisionTreeClassifier()

# Explanation:
# The DecisionTreeClassifier() initializes a decision tree model.
# The decision tree is a supervised learning algorithm used for classification tasks.
# In this case, it will be used to predict the 'Survived' column based on the selected features (X).

# Train the model with the training data
model.fit(X, y)

# Explanation:
# The fit() method is used to train the model.
# It takes the feature data (X) and the target data (y) as inputs.
# The model learns the patterns in the data and creates a decision tree that can predict survival based on the features provided.
```
